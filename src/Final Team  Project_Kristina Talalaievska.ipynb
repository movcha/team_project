```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
data = pd.read_csv('/Users/kristina/python/Final project/mortgage.csv')

data['year'] = pd.to_datetime(data['year'], format='%Y')

# Aggregate loan amounts by year
loan_by_year = data.groupby(data['year'].dt.year)['loan_amount'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=loan_by_year, x='year', y='loan_amount', marker='o')

# Customizing the plot
plt.title('Growing Confidence in the Lending Market: 2006-2016')
plt.xlabel('Year')
plt.ylabel('Total Loan Amount')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
data = pd.read_csv('/Users/kristina/python/Final project/mortgage.csv')

# Aggregate loan amounts by region
loan_by_region = data.groupby('region')['loan_amount'].sum().reset_index()

# Sort data for better visualization
loan_by_region = loan_by_region.sort_values(by='loan_amount', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=loan_by_region, x='region', y='loan_amount', palette='viridis')

# Customizing the plot
plt.title('Identifying Regions Higher on Loan Amounts and Showing More Economic Growth')
plt.xlabel('Region')
plt.ylabel('Total Loan Amount')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.show()
```
